## Relational Graphs
Let's first **def**ine a couple of functions, in both the mathematical and pythonic manner.

```python
def sin2(x):   #in our assignments, we have to use this function.
    """
    Calculates sin^2 of x
    """
    f = np.sin(x) ** 2
    return f
```

```python
def circle():
    """
    Returns x and y values of the unit circle
    """
    phi = np.linspace(0.0, 2.0 * np.pi, 1000)
    x = np.cos(phi)
    y = np.sin(phi)
    return x, y
```

Note, it is good practice to follow the order: imports -> function/class definitions -> working script.
In addition, functions should always have a docstring which explains what they do, and what each parameter does.
In a script, python functions should be seperated from each other by two lines, unless they are in a class, in which case it should be one line.
As we're in a notebook, good practice is to have a single cell for a single function.

```python
# calculate the sine of ONE value
x = 2.0
y = np.sin(x)
print("y = ", y)

# create an array with 1000 x values from -10 to 10 (inclusive) and printing first 10
# hint: try also using np.arange, instead.
x = np.linspace(-10.0, 10.0, 1000)
print("x =", x[:10])

# calculate the sine for all values at once
y = np.sin(x)
print("y =", y[:10])
```

```python
def plot_sine_cos(x, sinx, cosx):
    """
    Plotting sine and cosine of x array as line graph
    """
    # initialise figure, dpi is the resolution (144 looks good on big screen)
    plt.figure(dpi=144)

    # plot line graph, labels are used to produce the legend
    plt.plot(x, sinx, label="sin(x)")
    plt.plot(x, cosx, label="cos(x)")

    # formatting
    # set the upper and lower limits of x
    plt.xlim(-10.0, 10.0)
    # add axis labels
    plt.xlabel("x")
    plt.ylabel("f(x)")
    # add the legend, which uses the labels
    plt.legend(loc="upper left")
    plt.show()
    return
```

```python
# "call" that function with the created values
plot_sine_cos(x, np.sin(x), np.cos(x))  #This is the main programe. it will ask the sub programme above to to do some work and return the work to the main one.
```

```python
# now with sin^2(x)
def plot_sine_sine2(x, sinx, sin2x):
    """
    Plotting sine and sin^2 of x array as line graph
    """
    # initialise figure, dpi is the resolution (144 looks good on big screen)
    plt.figure(dpi=144)

    # plot line graph, labels are used to produce the legend
    plt.plot(x, sinx, label="sin(x)")
    plt.plot(x, sin2x, label="sin2(x)")

    # formatting
    # set the upper and lower limits of x
    plt.xlim(-10.0, 10.0)
    # add axis labels
    plt.xlabel("x")
    plt.ylabel("f(x)")
    # add the legend, which uses the labels
    plt.legend(loc="upper left")
    plt.show()
    return
```

```python
plot_sine_sine2(x, np.sin(x), sin2(x))
```

Note the differences!
Also a point to be wary, the third argument in our new function has a very similar name, "sin2x", to the function we defined at the start "sin2".
Be very careful not to override any variables, this is known as namespace contamination.
Instead of using plt.show, one can also use plt.savefig('filename.extension') to save the file directly.

# Question 1
Recreate the sin and cos graph but using sinh and cosh functions in place (see np.sinh and np.cosh).

```python
def plot_hyperbolic_sine_cos(x, sinhx, coshx):
    """
    Plotting hyperbolic sine and cosine of x array as line graph
    """
    # initialise figure, dpi is the resolution (144 looks good on big screen)
    plt.figure(dpi=144)

    # plot line graph, labels are used to produce the legend
    plt.plot(x, sinhx, label="sinh(x)")
    plt.plot(x, coshx, label="cosh(x)")

    # formatting
    # set the upper and lower limits of x
    plt.xlim(-10.0, 10.0)
    plt.ylim(-10, 10)
    # add axis labels
    plt.xlabel("x")
    plt.ylabel("f(x)")
    # add the legend, which uses the labels
    plt.legend(loc="upper left")
    plt.show()
    return
```

```python
plot_hyperbolic_sine_cos(x, np.sinh(x), np.cosh(x))
```

```python
def plot_circle(x, y):
    """
    Plots the unit circle for given x y positions
    """
    plt.figure(dpi=144)
    # plot line graph
    plt.plot(x, y)

    # formatting
    # add axis labels
    plt.xlabel("x")
    plt.ylabel("y")
    # axis equal sets the axis to look square
    plt.axis('equal')
    plt.show()
    return
```

```python
# let's "unpack" the variables from the circle function
x, y = circle()
plot_circle(x, y)

# alternatively, plot_circle(*circle()) also unpacks
```

# Question 2
Plot the trigonomical polynomials of cos(at) - cos(bt)^j against sin(ct) - sin(dt)^k, where:
<ul>
    <li>t = 0 to 2 * pi (try a different number of steps between those values!)</li>
    <li>a = 1</li>
    <li>b = 60</li>
    <li>c = 1</li>
    <li>d = 120</li>
    <li>j = 3</li>
    <li>k = 4</li>
</ul>

```python
def trigon1(a, b, j, t):
    """
    Computes the trigonomical polynomial cos(at) - cos(bt)**j with t from 0.0 to 2*pi
    """
    f = np.cos(a * t) - np.cos(b * t) ** j
    return f
```

```python
def trigon2(c, d, k, t):
    """
    Computes the trigonomical polynomial cos(ct) - cos(dt)**k with t from 0.0 to 2*pi
    """
    f = np.sin(c*t) - np.sin(d*t)**k
    return f
```

```python
def plot_trigon(trig1, trig2):
    """
    Plots the trigonomical polynomials against each other
    """
    plt.figure(dpi=144)

    plt.plot(trig1, trig2)

    # formatting
    plt.xlabel('Trig Poly 1')
    plt.ylabel('Trig Poly 2')
    plt.show()
    return
```

```python
a = 1
b = 60
c = 1
d = 120
j = 3
k = 4
t = np.linspace(0, 2 * np.pi, 10000)
trig1 = trigon1(a, b, j, t)
trig2 = trigon2(c, d, k, t)
plot_trigon(trig1, trig2)
```

Scatter plots are created in a much similar manner

```python
# creating random data from 0 to 1
x = np.random.uniform(0, 1, 100)
# creating "noisy" data of the same size
y = x + np.random.normal(0, 0.2, len(x))
```

```python
def plot_random_scatter(x, y):
    """
    Creates a scatter plot of random x and y data
    """
    plt.figure(dpi=144)

    # scatter plot
    plt.scatter(x, y, color='blue')
    # another trick: try instead plt.plot(x, y, 'bo')

    # formatting
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    return
```

```python
plot_random_scatter(x, y)
```

# Question 3
Considering we know the data should go from 0 to 1, copy the random scatter plot function above and add an overplotted straight black line.
We'll do a better job "fitting" this later in the module!

```python
def plot_random_scatter_with_line(x, y, xline, yline):
    """
    Creates a scatter plot of random x and y data with overplotted line
    """
    plt.figure(dpi=144)

    # scatter plot
    plt.plot(x, y, 'bo')
    plt.plot(xline, yline, 'k-')

    # formatting
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    return
```

```python
# creating random data from 0 to 1
x = np.random.uniform(0, 1, 100)
# creating "noisy" data of the same size
y = x + np.random.normal(0, 0.2, len(x))

# line data
xline = np.linspace(0, 1, 10)
yline = xline
plot_random_scatter_with_line(x, y, xline, yline)
```
